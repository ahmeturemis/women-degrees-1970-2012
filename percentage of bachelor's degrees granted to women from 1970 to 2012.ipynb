```python
import pandas as pd
import matplotlib.pyplot as plt

women_degrees = pd.read_csv('percent-bachelors-degrees-women-usa.csv')
x_values = women_degrees["Year"]
y_values = women_degrees["Biology"]
plt.plot(x_values,y_values)   #basic plot creating
plt.show()
```

```python
fig = plt.figure()
plt.plot(women_degrees["Year"],women_degrees["Biology"],c="blue",label="Women")
plt.plot(women_degrees["Year"],100-women_degrees["Biology"],c="green",label="Men")
plt.title("Percentage of Biology Degrees Awarded By Gender")
plt.legend(loc="upper right")
plt.show()
```

```python
fig = plt.figure()
plt.plot(women_degrees["Year"],women_degrees["Biology"],c="blue",label="Women")
plt.plot(women_degrees["Year"],100-women_degrees["Biology"],c="green",label="Men")
plt.tick_params(bottom="off",top="off",left="off",right="off")  #removed tick marks
plt.title("Percentage of Biology Degrees Awarded By Gender")
plt.legend(loc="upper right")
plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(women_degrees['Year'], women_degrees['Biology'], label='Women')
ax.plot(women_degrees['Year'], 100-women_degrees['Biology'], label='Men')
ax.tick_params(bottom="off", top="off", left="off", right="off")

ax.spines["right"].set_visible(False)        #remove spines
ax.spines["left"].set_visible(False)
ax.spines["bottom"].set_visible(False)
ax.spines["top"].set_visible(False)
ax.legend(loc='upper right')
ax.set_title('Percentage of Biology Degrees Awarded By Gender')
plt.show()
```

```python
stem_cats = ['Engineering', 'Computer Science', 'Psychology', 'Biology', 'Physical Sciences', 'Math and Statistics']
lib_arts_cats = ['Foreign Languages', 'English', 'Communications and Journalism', 'Art and Performance', 'Social Sciences and History']
other_cats = ['Health Professions', 'Public Administration', 'Education', 'Agriculture','Business', 'Architecture']
cb_dark_blue = (0/255,107/255,164/255)
cb_orange = (255/255, 128/255, 14/255)
     
```

```python
fig = plt.figure(figsize=(16,16)) #create an empty plot
#first column
for sp in range(0,18,3):   #range from 0 to 18 with +3 counting 
    col_co = int(sp/3)
    ax = fig.add_subplot(6,3,sp+1)  #create subplot has 6 rows
    ax.plot(women_degrees['Year'], women_degrees[stem_cats[col_co]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[stem_cats[col_co]], c=cb_orange, label='Men', linewidth=3)
    for key,spine in ax.spines.items(): #remove spines
        spine.set_visible(False)
    ax.set_xlim(1968, 2011) #limits x axis
    ax.set_ylim(0,100)      #limits y axis
    ax.set_yticks([0,100])  #create ticks in list
    ax.axhline(50, c=(171/255, 171/255, 171/255), alpha=0.3) #put a transparant line to 50 on y axis
    ax.set_title(stem_cats[col_co])
    ax.tick_params(bottom="off", top="off", left="off", right="off",labelbottom='off') #labelbottom remove line chart 
    if col_co == 0:
        ax.text(2005,87,"Men")     #ax.text adds text to plot. ax.text(x,y,string)
        ax.text(2002,8,"Women")
    elif col_co == 5:
        ax.text(2005,62,"Men")
        ax.text(2001,35,"Women")
        ax.tick_params(labelbottom='on') 


#second column
for sp in range(1,16,3):
    col_co = int((sp-1)/3)
    ax = fig.add_subplot(6,3,sp+1)
    ax.plot(women_degrees['Year'], women_degrees[lib_arts_cats[col_co]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[lib_arts_cats[col_co]], c=cb_orange, label='Men', linewidth=3)
    for key,spine in ax.spines.items():
        spine.set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_yticks([0,100])
    ax.axhline(50, c=(171/255, 171/255, 171/255), alpha=0.3)
    ax.set_title(lib_arts_cats[col_co])
    ax.tick_params(bottom="off", top="off", left="off", right="off",labelbottom='off')
    if col_co == 0:
        ax.text(2005,80,"Women")     #ax.text adds text to plot. ax.text(x,y,string)
        ax.text(2006,20,"Men")
    elif col_co == 4:
        ax.tick_params(labelbottom='on')

#third column 
for sp in range(2,20,3):
    col_co = int((sp-2)/3)
    ax = fig.add_subplot(6,3,sp+1)
    ax.plot(women_degrees['Year'], women_degrees[other_cats[col_co]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[other_cats[col_co]], c=cb_orange, label='Men', linewidth=3)
    for key,spine in ax.spines.items():
        spine.set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_yticks([0,100])
    ax.axhline(50, c=(171/255, 171/255, 171/255), alpha=0.3)
    ax.set_title(other_cats[col_co])
    ax.tick_params(bottom="off", top="off", left="off", right="off",labelbottom='off')
    if col_co == 0:
        ax.text(2005,90,"Men")     #ax.text adds text to plot. ax.text(x,y,string)
        ax.text(2004,2,"Women")
    elif col_co == 5:
        ax.text(2006,62,"Men")
        ax.text(2005,30,"Women")
        ax.tick_params(labelbottom='on')
plt.savefig("gender_degrees.png") #save plot as an image like .png
plt.show()
```
